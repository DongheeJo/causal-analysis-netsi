```python
from matplotlib import pyplot as plt
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pandas as pd
```

```python
#heteroskedasticity example
data = []
n = 100
for i in range(1,n):

    normal= np.random.normal(0, i, 10)        
    for norm in normal:
        data.append((i, norm))
        
x,y = zip(*data)
plt.scatter(x,y) 
```

How does heteroskedasticity affect our standard errors? (and in extension, the significance of our coefficients?)

Let's consider a data-generating process where variance in y increases when x is far from x's mean. My intuition is that the typical standard errors are too big, and the regression will have a hard time detecting significance.


```python
np.random.seed(10)
data = []
n = 100
slope = 2
for i in range(1,n):
    if i <=(n/2):
        normal= np.random.normal(slope*i, i**2, 10) #note positive slope here

    else:
       
        normal= np.random.normal(slope*i, (n-i)**2, 10)
    
    for norm in normal:
        data.append((i, norm))
        
x,y = zip(*data)
plt.scatter(x,y) 
```

```python
df = pd.DataFrame(data)
results = smf.ols('y ~ x', data=df).fit()
print(results.summary())
```

```python
results = smf.ols('y ~ x', data=df).fit(cov_type='HC3')
print(results.summary())
```

```python
#robust standard errors definitely increase significance
#it's a bit confusing
```

```python
np.random.seed(10)
data = []
n = 100
for i in range(1,n):
    if i <=(n/2):
        normal= np.random.normal(0, ((n/2)-i)**2, 10)
    else:
        normal= np.random.normal(0, (i-(n/2))**2, 10)
        
    for norm in normal:
        data.append((i, norm))
        
x,y = zip(*data)
plt.scatter(x,y) 
```

```python
#in this case, y's variance is higher when age is far from it's mean
#so we expect that the typical standard errors are too small
```

```python
df = pd.DataFrame(data)
results = smf.ols('y ~ x', data=df).fit()
print(results.summary())
```

```python
#this regression table tells us there is a weak correlation between age and y
#but we know that's not true, since we set the data-generating process to have mean 0
```

```python
results = smf.ols('y ~ x', data=df).fit(cov_type='HC3')
print(results.summary())
```

```python
#correcting for standard errors helps us realize the weakness of the correlation
```
